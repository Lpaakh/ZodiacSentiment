```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from clean_tweet import clean_tweet
```

```python
#VADER sentiment analyzer is a pre-trained sentiment analyzer designed for social media
analyser = SentimentIntensityAnalyzer()
```

```python
data = pd.read_csv('tweets_11_7.csv')
```

```python
data.sample(10)
```

```python
data['clean'] = data['text'].map(clean_tweet)
```

```python
data.head()
```

```python
def get_score(tweet):
    return analyser.polarity_scores(tweet)
```

```python
scores = list(data['clean'].map(get_score).values)
```

```python
scores = pd.DataFrame.from_dict(scores)
```

```python
scores
```

```python
data = data.join(scores)
```
